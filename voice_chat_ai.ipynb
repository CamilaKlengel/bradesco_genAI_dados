# 🎤 PASSO 2 — Gravação de Áudio Com Python (e Uma Pitada de JavaScript)

```python
from IPython.display import Audio, display, Javascript
from google.colab import output
from base64 import b64decode

# Código JavaScript para gravar áudio do usuário usando a "MediaStream Recording API"
RECORD = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=5):
  # Executa o código JavaScript para gravar o áudio
  display(Javascript(RECORD))
  # Recebe o áudio gravado como resultado do JavaScript
  js_result = output.eval_js('record(%s)' % (sec * 1000))
   # Decodifica o áudio em base64
  audio = b64decode(js_result.split(',')[1])
  # Salva o áudio em um arquivo
  file_name = 'request_audio.wav'
  with open(file_name, 'wb') as f:
    f.write(audio)
  # Retorna o caminho do arquivo de áudio (pasta padrão do Google Colab)
  return f'/content/{file_name}'

# Grava o áudio do usuário por um tempo determinado (padrão 5 segundos)
print('Ouvindo...\n')
record_file = record()

# Exibe o áudio gravado
display(Audio(record_file, autoplay=False))
```

# 🧠 PASSO 3 - Reconhecimento de Fala com Whisper (OpenAI)

```python
import whisper

# Selecione o modelo do Whisper que melhor atenda às suas necessidades:
# https://github.com/openai/whisper#available-models-and-languages
model = whisper.load_model("small")

# Transcreve o audio gravado anteriormente.
result = model.transcribe(record_file, fp16=False, language=language)
transcription = result["text"]
print(transcription)
```

# 🔑 PASSO 4 — Conectar com a IA (SUBSTITUTO GRATUITO)

```python
def gerar_resposta(texto):
    texto = texto.lower()

    contexto_ia = [
        "inteligência artificial", "ia", "machine learning", "deep learning"
    ]

    tem_contexto_ia = any(palavra in texto for palavra in contexto_ia)

    if tem_contexto_ia and "desvantagens" in texto:
        return "As desvantagens da inteligência artificial incluem dependência tecnológica, possíveis erros e preocupações com privacidade."

    elif tem_contexto_ia and "vantagens" in texto:
        return "As vantagens da inteligência artificial incluem automação de tarefas, aumento de produtividade e análise rápida de dados."

    elif tem_contexto_ia and "futuro" in texto:
        return "O futuro da inteligência artificial envolve sistemas cada vez mais inteligentes e presentes no dia a dia."

    elif tem_contexto_ia and "para que serve" in texto:
        return "A inteligência artificial serve para automatizar tarefas, analisar dados e ajudar na tomada de decisões."

    elif "inteligência artificial" in texto:
        return "Inteligência artificial é uma tecnologia que permite que máquinas aprendam e realizem tarefas que normalmente exigiriam inteligência humana."

    elif "machine learning" in texto or "aprendizado de máquina" in texto:
        return "Machine learning é uma área da inteligência artificial que permite que sistemas aprendam a partir de dados, melhorando seu desempenho sem serem programados explicitamente."

    elif "deep learning" in texto:
        return "Deep learning é uma técnica avançada de inteligência artificial que utiliza redes neurais para aprender padrões complexos, como reconhecimento de imagem e voz."

    elif "o que é python" in texto:
        return "Python é uma linguagem de programação muito utilizada em ciência de dados, inteligência artificial e desenvolvimento de sistemas."

    elif "oi" in texto or "olá" in texto:
        return "Olá! Como posso te ajudar hoje?"

    else:
        return f"Você perguntou sobre '{texto}'. Ainda estou aprendendo, mas posso te ajudar com esse tema."
```

# 🔊 PASSO 5 — Transformar resposta em voz (gTTS)

```python
resposta = gerar_resposta(transcription)

print("Resposta:", resposta)

from gtts import gTTS
from IPython.display import Audio

tts = gTTS(text=resposta, lang=language, slow=False)

audio_file = "resposta.mp3"
tts.save(audio_file)

Audio(audio_file, autoplay=True)
```
